```
# step1 merge the two movies and credit dataset and convert into one based on title
# Drop conflicting columns from 'credits' DataFrame before merging
#credits = credits.drop(columns=['crew', 'movie_id', 'cast'], errors='ignore')  # errors='ignore' to avoid error if column doesn't exist

movies = pd.merge(movies, credits, on='title')
```

```
#genres
#id
#keywords
#title
#overview
#cast
#crew

movies = movies[['movie_id','title','overview','genres','keywords','cast','crew',]]
```

```
import ast # Import the ast module

def convert(obj):
  L = []
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L # Return the list after the loop completes

movies['genres']=movies['genres'].apply(convert)
```

```
movies.head(1)
```

```
movies['keywords']=movies['keywords'].apply(convert)
```

```
movies.head(1)
```

```
import ast # Import the ast module

def convert(obj):
  L = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter != 3:
      L.append(i['name'])
      counter+=1
    else:
      break
  return L # Return the list after the loop completes

movies['cast']=movies['cast'].apply(convert)
```

```
movies.head(1)
```

```
def fetch_director(obj):
  L= []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      L.append(i['name'])
      break
  return L

movies['crew']=movies['crew'].apply(fetch_director)
```

```
movies.head(1)
```

```
movies['overview'][0]
```

```
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```
movies.head(1)
```

```
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```
movies.head(1)
```

```
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```
movies.head()
```

```
 new_df = movies[['movie_id','title','tags']]
```

```
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```
new_df.head()
```

```
import nltk
```

```
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```
def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```
new_df['tags']=new_df['tags'].apply(stem)
```

```
new_df['tags'][0]
```

```
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```
new_df.head()
```

```
new_df.shape
```

```
new_df['tags'][0]
new_df['tags'][1]
```

```
#  text vectorization of tag column (by using vector concept means convert the all movies in vector find the distance between movies  )
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
cv = CountVectorizer(max_features=1000000,stop_words='english')

# Calculate the vectors and store them in the 'vectors' variable
vectors = cv.fit_transform(new_df['tags']).toarray()

# Calculate the cosine similarity matrix using the 'vectors' array
similarty = cosine_similarity(vectors)

# Print the shape of the cosine similarity matrix
print(similarty.shape)
```

```
similarty[0]
```

```
sorted(list(enumerate(similarty[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances = similarty[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:10]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)

```

```
recommend("The Dark Knight Rises")
```
